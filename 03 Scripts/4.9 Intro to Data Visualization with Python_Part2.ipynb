### 2. Creating Bar Charts<a class="anchor" id="cbarc"></a>

```python
ords_prods_all['orders_day_of_week'].value_counts().plot.bar(title = 'Orders by day of the week', xlabel = 'Day numbers', ylabel = '# of Orders');
```

### 5. Line Charts <a class="anchor" id="linec"></a>

##### Sampling Data <a class="anchor" id="samp"></a>

```python
# Create a list holding True/False values to the test np.random.rant() <= 0.7
np.random.seed(4)
dev = np.random.rand(len(ords_prods_all)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big
big = ords_prods_all[dev] # True
```

```python
# Store 30% of the sample in the dataframe small
small = ords_prods_all[~dev]  # False
```

```python
len(ords_prods_all)
```

```python
len(big) + len(small)
```

```python
# Speed up the processing, use columns necessary for the chart: â€œpricesâ€ and â€œorders_day_week"
df_2 = small[['orders_day_of_week','prices']]
```

##### Create line chart <a class="anchor" id="clinec"></a>

```python
# Create line chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```
