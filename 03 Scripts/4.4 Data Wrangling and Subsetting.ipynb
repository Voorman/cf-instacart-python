```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

### 1. Importing data<a class="anchor" id="imp"></a>

```python
# Load order and product data

# set path 
path = r'/Users/ivarvoorman/CareerFoundry/Data Analytics/Data Immersion/4 Python Fundamentals for Data Analysts/01-2022 Instacart Basket Analysis'
 
# read orders.csv to dataframe df_ords
df_ords = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'orders.csv'),index_col = False)

# Load products.csv to dataframe df_prods
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'products.csv'),index_col = False)

# Load departments.csv to dataframe df_dep
df_dep = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'departments.csv'),index_col = False)
```

### 02. Data wrangling<a class="anchor" id="dwl"></a>

```python
# Dropping eval_set column from orders.csv
df_ords = df_ords.drop(columns = ['eval_set'])
```

```python
# print all the functions within column days_since_prior_order
df_ords['days_since_prior_order'].value_counts(dropna = False)
```

```python
# Change the name of column 'order_dow' to 'orders_day_of_week'
df_ords.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = True)
df_ords.head()
```

```python
# Changes the data type of the "order_id" column from a number to a string
df_ords['order_id'] = df_ords['order_id'].astype('str')
df_ords['order_id'].dtype
```

```python
# Transposing df.dep
df_dep_t =  df_dep.T
df_dep_t
```

```python
# Add an index to dataframe df_dep_t
df_dep_t.reset_index()
```

```python
# Create a new header
new_header = df_dep_t.iloc[0]

# Remove the first row in the dataframe
df_dep_t_new = df_dep_t[1:]

# Add the new header to the dataframe
df_dep_t_new.columns = new_header
df_dep_t_new
```

```python
# Turn df_dep_t_new dataframe into a dictionary
data_dict = df_dep_t_new.to_dict('index')
data_dict
```

### 03. Data Dictionaries<a class="anchor" id="dd"></a>

```python
# Use data dictionary to determine what "department_id" of 19 revers to
print(data_dict.get('19'))
```

### 04. Subsetting<a class="anchor" id="sub"></a>

```python
# Subset df_prods dataframe with only data from the snacks department
df_snacks =  df_prods[df_prods['department_id']==19]
df_snacks.head()
```

```python
# Subset df_prods dataframe with only data from the snacks department with loc
df_snacks_2 = df_prods.loc[df_prods['department_id'] == 19]
df_snacks_2.head()
```

```python
# Subset df_prods dataframe with only data from the snacks department with loc and isin
df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([19])]
df_snacks_3.head()
```
